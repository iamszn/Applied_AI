# Sales Analysis

```python
import pandas as pd
import os#to list all file of a diretory
```

# Task 1:Merging 12 months of sales data into a single file

```python
df=pd.read_csv('./Sales_Data/Sales_April_2019.csv')
files=[file for file in os.listdir('./Sales_Data') ]
all_months_data=pd.DataFrame()

# os.listdir list all file of a directory
for file in files:
    df=pd.read_csv('./Sales_Data/'+file)#df reads the data from each csv file
    all_months_data=pd.concat([all_months_data,df])
# all_months_data.head()
all_months_data.to_csv('all_months_data.csv',index=False)
```

# Read in updated format

```python
all_data=pd.read_csv('all_months_data.csv')
all_data.head()                  
```

# Clean Up Data
*  Drop NaN rows
*  Find Or and Delete it
*  Converts column Quantity orderd and Price Each to correct type

```python
nan_df=all_data[all_data.isna().any(axis=1)]
# Find all the columns with nan value
nan_df.head()
all_data=all_data.dropna()
# removes all the rows having nan value
all_data.head()
```

```python
tmp_df=all_data[all_data['Order Date'].str[0:2]=='Or']
# finds all the rows having improper order date
tmp_df.head()
```

```python
all_data=all_data[all_data['Order Date'].str[0:2]!='Or']
```

```python
all_data['Price Each']=pd.to_numeric(all_data['Price Each'])#to int
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered'])#to float
```

# Augment data with additional columns

# Task 2:Add month columns

```python
all_data['Month']=all_data['Order Date'].str[0:2]
# all_data['Month']=all_data['Month'].astype('int32')
all_data
```

# Task 3:Add a Sales Columns

```python
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

## Task 4:Add a city Column

```python
# use .apply()
def get_city(address):
    return address.split(',')[1]
#     get only city name from Purchase address

def get_state(address):
    return address.split(',')[2].split(' ')[1]
#     get only state name from Purcahse Address
    

# all_data=all_data.drop(columns='Column')
all_data['City']=all_data['Purchase Address'].apply(lambda x:f"{get_city(x)}({get_state(x)})")
all_data.head()
```

### Question 1:What was the best month for sales?How much was earned that month

```python
results=all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt
months=range(1,13)
plt.bar(months,results['Sales'])
plt.xticks(months)
plt.xlabel('Months in Number')
plt.ylabel('Sales in USD($)')
```

### Question 2:What city had the highest no of sales?

```python
results=all_data.groupby('City').sum()
results
```

```python
import matplotlib.pyplot as plt
cities=[city for city,df in all_data.groupby('City')]
plt.bar(cities,results['Sales'])
plt.xticks(cities,rotation='vertical',size=8)
plt.xlabel('Cities Name')
plt.ylabel('Sales in USD($)')
```

## Question 3:What time should we display advertisement to maximize likelihood of customer's buying product?  

```python
all_data['Order Date']=pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour']=all_data['Order Date'].dt.hour
all_data['Hour']=all_data['Order Date'].dt.hour
all_data.head()
```

```python
hours=[hour for hour,df in all_data.groupby('Hour')]
plt.plot(hours,all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.grid()
# all_data.groupby(['Hour']).count()
plt.show()
# hours vs frequency of hours[no of products ordered;how many products were ordered in this hour]
```

## Question 4:What products are most sold together?
*  [HINT]Finding product with same order ID

```python
df=all_data[all_data['Order ID'].duplicated(keep=False)]#Mark all duplicates as ``True``.
df.head()
```

```python
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
# creates new column Grouped and store products name having same order date
df.head()
```

```python
df=df[['Order ID','Grouped']].drop_duplicates()
#removes all duplicate values based on order id and  grouped
df.head()
```

```python
from itertools import combinations
from collections import Counter

count=Counter()
for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
#     count all products as a pair(2 products combination frequency)
# print(count)
for key,value in count.most_common(10):
    print(key,value)
```

## Question 5:What product sold the most ?Why do you think it sold the most?

```python
all_data.head()
```

```python
product_group=all_data.groupby('Product')
# product_group.sum()
quantity_ordered=product_group.sum()['Quantity Ordered']

products=[product for product,df in product_group]
plt.bar(products,quantity_ordered)
plt.xticks(products,rotation='vertical',size=14)
plt.xlabel('Products')
plt.ylabel('Quantity Order')
plt.show()
```

```python
prices=all_data.groupby('Product').mean()['Price Each']
print(prices)
```

```python
prices=all_data.groupby('Product').mean()['Price Each']
fig,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.bar(products,quantity_ordered,color='g')
ax2.plot(products,prices,'r-')

ax1.set_xlabel('Product Name',color='r')
ax1.set_ylabel('Quantity Ordered',color='g')
ax2.set_ylabel('price($)',color='b')
ax1.set_xticklabels(products,rotation='vertical',size='14')
plt.show()
```
