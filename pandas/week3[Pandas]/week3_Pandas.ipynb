## Imputation
 
Imputation is a conventional feature engineering technique used to keep valuable data that have null values.

There may be instances where dropping every row with a null value removes too big a chunk from your dataset, so instead we can impute that null with another value, usually the mean or the median of that column.


```python
new_movie=pd.read_csv('IMDB_movie.csv',index_col='Title')
new_movie.head()
```

```python
new_movie.isnull().any()
```

```python
new_movie.isnull().sum()
```

```python
revenue_mean=new_movie['Revenue (Millions)'].mean()
metascore_mean=new_movie['Metascore'].median()
```

```python
new_movie['Revenue (Millions)'].fillna(revenue_mean,inplace=True)
new_movie['Metascore'].fillna(metascore_mean,inplace=True)
```

 When inplace = True is used, it performs operation on data and nothing is returned. When inplace=False is used, it performs operation on data and returns a new copy of data

## Applying Functions:
*  apply function to convert movies with rating 8.0 or greater to a string value of "good" and the rest to "bad" and use this transformed values to create a new column.

* The `.apply()` method passes every value in the `Rating` column through the `rating_function` and then returns a new Series. This Series is then assigned to a new column called `rating_category`.
 

```python
def rating_function(x):
    if x >= 8.0:
        return "good"
    else:
        return "bad"

    
```

```python
new_movie["rating_category"] = new_movie["Rating"].apply(rating_function)

new_movie.head(2)
```

## Lambda
* lambda performs the same operation as above function

```python
new_movie["rating_category"] = new_movie["Rating"].apply(lambda x: 'good' if x >= 8.0 else 'bad')

new_movie.head(2)
```
