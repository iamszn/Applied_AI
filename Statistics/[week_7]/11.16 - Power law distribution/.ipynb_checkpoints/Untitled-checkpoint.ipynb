## Discussion:
* To check if the two variable X=(x1,x2....xn) and Y=(y1,y2,y3....yn) follow power law or not we can use:
* -chebyshev Distribution
* OR
* calculate the log of each pair as Xnew=log(x1),log(x2)....log(xn) and Ynew=log(y1),log(y2)....log(yn) and plot the new value
* If the straight line is formed as shown in the figure then they follow power law
