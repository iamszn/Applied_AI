## CLT[Central Limit Theorem]
* used to estimate the mean and variance of total population using mean and variance of sampling distribution of sample means 
* 1.From the given population first m no of samples of size n is taken
* 2.Mean of all sample(from sample 1 to sample m) is calculated
* 3.Now Gaussian distribution of sample means(mean of each samples) is calculated
* 4.Using the mean and variance  of sample means original mean and variance of the totoal population is calculated
