# Q-Q Plot
* QQ plot is used to determine whether a given distribution is normally distributed or not
* If X and Y are two random distributions then QQ plot is used to find wheteher they have same distribution or not

```python
#Q-Q plot
import numpy as np 
import pylab 
import scipy.stats as stats

# N(0,1)
std_normal = np.random.normal(loc = 0, scale = 1, size=1000)
# generate samples rndomly from gaussian/normal distribution 
# with mean(loc=0) and variance(scale=1) and generate 1000s such samples

# 0 to 100th percentiles of std-normal
for i in range(0,101):
    print(i, np.percentile(std_normal,i))
#     print the value of each percentile from 0 to 100th
# and their corresponding value from the normal distribution


```

```python
# generate 100 sanples from N(20,5)
measurements = np.random.normal(loc = 20, scale = 5, size=50) 
#try size=1000


stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
# display the Q-Q or probability plot by comparing
# random distribution having mean 20 and variance 5 
# with normal distribution having mean 0 and variance 1
```

## Discussion:
* If no of sample is small then its hard to intrepret QQ plot 
* QQ plot dont work on small no of samples
    

```python
# generate 100 sanples from N(20,5)
measurements = np.random.normal(loc = 20, scale = 5, size=50000) 
#try size=1000


stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

## Discussion:
* QQ plot works effectively if the no of samples are high

```python
# generate 100 sanples from N(20,5)
measurements = np.random.uniform(low=-1, high=1, size=100) 
#try size=1000

stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

## Discussion:
* If all the points lie on the straight line then distribution on x axis and distribution on y axis lie on the same family
* If points are deviated and dont lie on the same family then distribution on x axis and distribution on y axis lie on different family
* In the above figure all points dont lie on the straight line so the two distributions are not of same family i.e.Gaussian distribution lies on the X axis and uniform distribution lies on the Y axis
