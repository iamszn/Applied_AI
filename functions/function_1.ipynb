# enumerate()

```python
from IPython.display import Image
Image(filename='enum.jpg')
```

```python
num=[2,4,6,8]
for index,no in enumerate(num):
    print('index {0} has value {1}'.format(index,no))
# default index is 0
```

```python
num=[2,4,6,8]
for index,no in enumerate(num,2):
    print('index {0} has value {1}'.format(index,no))
# here the index start from 2
```

# filter()

```python
def find_positive_number(n):
    if n>0:
        return n
    
num_list=range(-10,10)
print(list(num_list))

positive_no_lst=list(filter(find_positive_number,num_list))
print('[INFO] Positive numbers are:\n',positive_no_lst)
```

```python
def find_even_num(no):
    if (no%2==0):
        return no
    
num_lst=[1,2,3,4,5,6,7,8,9,0]
print(num_lst)

even_no=list(filter(find_even_num,num_lst))
print('[INFO] Even numbers are:\n',even_no)
# filter is function that filters the elements of list according
# to the given conditions
```

# map()

```python
from IPython.display import Image
Image(filename='map.jpg')
```

```python
numbers=[1,2,3,4]
def Cube(n):
    return(n**3)

# using map() function    
cube_no=list(map(Cube,numbers))
print('[INFO] Cube of numbers are:\n',cube_no)
```

# Simple Calculator:

```python
def add(a,b):
    return(a+b)
def subtract(a,b):
    return(a-b)
def multiply(a,b):
    return(a*b)
def division(a,b):
    return(a/b)

print('[INFO]Select Option')
print('1.Addition')
print('2.Subtraction')
print('3.Multiplication')
print('4.Division')

ip=int(input('Enter choice 1/2/3/4'))
num1=float(input('Enter first no:'))
num2=float(input('Enter second no:'))

if ip==1:
    print('[INFO]:Addition of {0} and {1} is:{2}\n'.format(add(num1,num2)))
elif ip==2:
    print('[INFO]:Subtraction of {0} and {1} is:{2}\n'.format(subtract(num1,num2)))
elif ip==3:
    print('[INFO]:Multiplication of {0} and {1} is:{2}\n'.format(multiply(num1,num2))) 
elif ip==4:
    print('[INFO]:Division of {0} and {1} is:{2}\n'.format(division(num1,num2)))    
else:
    print('[INFO]:Invalid choice')
```
