## CDF
* Using CDF percentile of a given value of x axis can be found in  y-axis

```python
status_1=df_haber.loc[df_haber['Surv_Status']==1]
# from dataframe df_haber find all the row whose with Surv_status value is 1
status_2=df_haber.loc[df_haber['Surv_Status']==2]
# from dataframe df_haber find all the row whose with Surv_status value is 1
```

```python
counts,bin_edges=np.histogram(status_1['Age'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf,label='CDF')
plt.xlabel('Age')
plt.title('[CDF plot 1.1]Age feature corresponding to class status_1')
plt.grid()
plt.legend()
plt.show()
```

```python
counts,bin_edges=np.histogram(status_2['Age'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf,label='CDF')
plt.xlabel('Age')
plt.title('[CDF plot 1.2]Age feature corresponding to class status_2')
plt.grid()
plt.legend()
plt.show()
```

```python
counts,bin_edges=np.histogram(status_1['Operation_Year'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf,label='CDF')
plt.xlabel('Age')
plt.title('[CDF plot 1.1]Operation_Year feature corresponding to class status_1')
plt.grid()
plt.legend()
plt.show()
```

```python
counts,bin_edges=np.histogram(status_2['Operation_Year'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf,label='CDF')
plt.xlabel('Age')
plt.title('[CDF plot 2.2]Operation_Year feature corresponding to class status_2')
plt.grid()
plt.legend()
plt.show()
```

```python
counts,bin_edges=np.histogram(status_1['Axil_Nodes'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf,label='CDF')
plt.xlabel('Age')
plt.title('[CDF plot 3.1]Axil_Nodes feature corresponding to class status_1')
plt.grid()
plt.legend()
plt.show()
```

```python
counts,bin_edges=np.histogram(status_2['Axil_Nodes'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf,label='CDF')
plt.xlabel('Age')
plt.title('[CDF plot 3.2]Axil_Nodes feature corresponding to class status_2')
plt.grid()
plt.legend()
plt.show()
```

## Bivariate Analysis

## Pair Plot

```python
# pairwise scatter plot: Pair-Plot
plt.close();
sns.set_style("whitegrid");
sns.pairplot(df_haber, hue="Surv_Status",x_vars =['Age','Operation_Year','Axil_Nodes'],y_vars =['Age','Operation_Year','Axil_Nodes'], size=4);
plt.suptitle("Pair-Plot of Haberman Data",size =26);
plt.show();
```
