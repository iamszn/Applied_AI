```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('iris.csv')
df.head()
```

```python
df.shape
```

## Univariate Analysis 
* In Univariate Analysis we take a single feature and analyze it.Like we can take one feature as sepal length and on the basis of it we can identify setosa,virginica or versicolor
* Univariate  Analysis helps to find the most important features from the list of other features

```python
df_setosa=df.loc[df['species']=='setosa']
# all the rows of species setosa
df_virginica=df.loc[df['species']=='virginica']
# all the rows of species virginica
df_versicolor=df.loc[df['species']=='versicolor']
# all the rows of species versicolor
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'*')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'+')
plt.xlabel('Sepal_length')
plt.title('1D-SCATTER PLOT')
plt.show()
```

```python
sns.FacetGrid(df,hue='species',size=5).map(sns.distplot,'petal_length').add_legend();
plt.grid()
plt.title('CASE-1:Histogram of petal_length')
plt.show()
```

```python
sns.FacetGrid(df,hue='species',size=5).map(sns.distplot,'petal_width').add_legend();
plt.grid()
plt.title('CASE-2:Histogram of petal_width')
plt.show()
```

```python
sns.FacetGrid(df,hue='species',size=5).map(sns.distplot,'sepal_length').add_legend();
plt.grid()
plt.title('CASE-3:Histogram of sepal_length')
plt.show()
```

```python
sns.FacetGrid(df,hue='species',size=5).map(sns.distplot,'sepal_width').add_legend();
plt.grid()
plt.title('CASE-4:Histogram of sepal_width')
plt.show()
```

## Conclusion:
* the distance between histogram of setosa,versicolor and virginica is greatest in petal length
* petal_length have the least overlapping of pdf and histogram
* petal_length gives the best separation
* PL>PW>>SL>>SW
