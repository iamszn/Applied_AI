## Plot using matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
# initialize figure and axes object
fig,ax=plt.subplots()
# load in data
tips=pd.read_csv('tips.csv')
# Create violinplot
ax.violinplot(tips["total_bill"], vert=False)# show the plot
plt.grid()
plt.show()
# tips.head()
```

* Matplotlib doesnt give the information about percentile

## Plot using Seaborn

```python
# import libraries
import matplotlib.pyplot as plt
import seaborn as sns

# load the data
tips=sns.load_dataset('tips')

# create violinplot
sns.violinplot(x='total_bill',data=tips)

# show the plot
plt.grid()
plt.show()
```

* Using seaborn we can get the information about percentile as shown in above fig through boxplot
